```python
import pandas as pd
import re

pathInfo='Datos.tsv'
```

```python
df=pd.read_csv(pathInfo, header=0, delimiter=r'\t')
```

```python
df
```

```python
val=df['Location'].tolist()
state=[]
for i in val:
    j=re.split('/',i)
    state.append(j[2])
df['Location']=state
```

```python
val=df['AA Substitutions'].tolist()
dicAA=[]
for i,j in enumerate(val):
    for x in re.split('[(,)"]',val[i-1]):
        if len(x) > 1:
            dicAA.append(x)
dicAA=set(dicAA)
```

```python
val=df['AA Substitutions'].tolist()
dfDicAA={}

for i in dicAA:
    columnDic=[]
    for k in df['AA Substitutions']:
        if i in k:
            columnDic.append(1)
        else:
            columnDic.append(0)
    dfDicAA[i]=columnDic
dfD=pd.DataFrame(dfDicAA)
```

```python
dfDatos=pd.concat([df,dfD], axis=1)
```

```python
dfDatos.to_csv('DatosAA.csv',encoding='utf-8')
```
